**Data:**


This dataset includes several months (and counting) of data on daily trending YouTube videos. Data is include for the US, GB, jp, and FR regions with up to 200 listed trending videos per day. in our project, we target our dataset with 40949 rows and 16 columns.

Through the data set, we extracted answers to our questions and represented them in the form of tables and charts.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 


```

**read data files**

```python
USvideo = pd.read_csv("../data/USvideos.csv")
GBvideo= pd.read_csv("../data/GBvideos.csv")
JPvideo = pd.read_csv("../data/JPvideos.csv")
FRvideo = pd.read_csv("../data/FRvideos.csv") 
```

```python
USvideo['country']='US'
GBvideo['country']='GB'
JPvideo['country']='JP'
FRvideo['country']='FR' 

```

```python
USvideo.drop_duplicates(subset=['video_id']) 
GBvideo.drop_duplicates(subset=['video_id'])
JPvideo.drop_duplicates(subset=['video_id'])
FRvideo.drop_duplicates(subset=['video_id']) 


```

**contain US,GB,JP and FR**

```python
ALLvideo =pd.concat([USvideo,GBvideo,JPvideo,FRvideo])
ALLvideo.drop_duplicates(subset=['video_id'])  
```

```python
ALLvideo['Popularity']=((ALLvideo['views'] * 0.75) + (ALLvideo['likes'] * 1) + (ALLvideo['comment_count'] * 1.5) - 
                        (ALLvideo['dislikes'] * 2)) / 100000

ALLvideo['Popularity']=ALLvideo['Popularity'].round(decimals=2)    
```

**#Top 1 Popular videos from each category** 

```python
top_cat=(ALLvideo[['category_id','country','channel_title','views','dislikes','comment_count','Popularity']]
             .sort_values('Popularity',ascending=False).groupby('category_id'))   
```

```python
top_cat=top_cat.head(1).sort_values(['category_id','Popularity'],ascending=[True,False]).copy()
top_cat.reset_index(drop=True,inplace=True) 
```

**# print top 5**

```python
def print_top5(column_of_interest,column_stats):
    USvideo.drop_duplicates(subset ="title",
                     keep = 'last', inplace = True)
    df = USvideo.groupby(column_of_interest)['views', 'likes','comment_count', 'dislikes'].apply(lambda x: x.astype(int).sum())
    return df.sort_values(by=column_stats, ascending=False).head() 
```

```python
print_top5('title','views')  
```
